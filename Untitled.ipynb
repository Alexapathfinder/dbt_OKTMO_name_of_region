Пролемматизируем столбец "Наименование" таблицы nas_new:

```python
%%time
m = Mystem()
purpose_lemmatized = []
for row in nas_new['Наименование']:
    lemmas = ' '.join(m.lemmatize(row)).strip()
    purpose_lemmatized.append(lemmas.split())
nas_new['Наименование_new'] = purpose_lemmatized
print(nas_new)
```
